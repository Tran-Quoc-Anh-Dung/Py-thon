```python
# ví dụ 1
# Bạn là chủ một hiệu sách tiếng Anh và đây là một số đầu sách kinh điển bạn đang bán:

classics = {"Austen": "Pride and Prejudice",

"Shelley":"Frankenstein",

"Joyce":"Ulyssessss"}

print(classics)

print("------------------------------------------------------------------------------------")

# ví dụ 2
# • Bạn đang tiến hành kiểm kê và cần in tên tác giả và tiêu đề:

# as dict_items

print(classics.items())

# như một danh sách các bộ

print(list(classics.items()))

print("------------------------------------------------------------------------------------")

# ví dụ 3
# • Sau đó, bạn cần in riêng tên tác giả và tiêu đề:

# authors as dict_items

print(classics.keys())

# authors as a list

print(list(classics.keys()))

print("------------------------------------------------------------------------------------")

# ví dụ 4 

# titles as dict_items.

print(classics.values())

#titles as a list

print(list(classics.values()))


print("------------------------------------------------------------------------------------")

# ví dụ 5
# • Bạn nhận thấy tiêu đề của cuốn sách cuối cùng bị sai nên bạn sửa lại:
# Cuốn sách cuối cùng tên tác giả là “Joyce”

print("Wrong title: " + classics ["Joyce"])

print("------------------------------------------------------------------------------------")

# ví dụ 6

classics ["Joyce"] = "Ulysses"

print("Corrected title: " + classics["Joyce"])

# ví dụ 7
# Sau đó, bạn thêm hai cuốn sách mới vừa ra mắt: Gulliver’s Travels của Swift và Jane Eyre của Bronte:

# adding the first book (syntax 1)

classics["Swift"] = "Gulliver's travels"

print(classics)

print("------------------------------------------------------------------------------------")

# ví dụ 8

# adding the second book (syntax 2)

classics.update({"Bronte": "Jane Eyre"})

print(classics)

print("------------------------------------------------------------------------------------")


# ví dụ 9
# • Cuối cùng, bạn loại bỏ những cuốn sách của Austen và Joyce vì bạn vừa bán chúng:

# deleting the first book (syntax 1)

del classics ["Austen"]

print(classics)


print("------------------------------------------------------------------------------------")


# ví dụ 10

# deleting the second book (syntax 2)

classics.pop("Joyce")

print(classics)
```

```python
# bài tập 1
#1. Thông tin sinh viên. Đối với tình huống sau, hãy tạo mã tương tự như mã được trình bày trong chương này. Bạn làm việc tại Phòng Đăng ký của một trường học và đây là dữ liệu của một sinh viên:
# student = {"First name":"Bruce", "Last name":"Zhiang", "Sex":"Male", "Age":21, "Course":"Literature", "Hobby":"Swimming"}

# Create the student dictionary
student = {"First name": "Bruce", "Last name": "Zhiang", "Sex": "Male", "Age": 21,
           "Course": "Literature", "Hobby": "Swimming"}

# Tạo từ điển student
student = {"First name": "Bruce", "Last name": "Zhiang", "Sex": "Male", "Age": 21,
           "Course": "Literature", "Hobby": "Swimming"}

# a. In ra tất cả các khóa và giá trị của chúng
print("a. Tất cả các khóa và giá trị:")
for key, value in student.items():  # Duyệt qua từng cặp khóa-giá trị trong từ điển
    print(f"{key}: {value}")
print("------------------------------------------------------------------------------------")

# b. In ra tất cả các khóa
print("b. Tất cả các khóa:")
print(student.keys())  # In ra đối tượng dict_keys chứa tất cả các khóa
# Cách khác: in ra dưới dạng danh sách
print(list(student.keys()))  # Chuyển đổi dict_keys thành list để hiển thị
print("------------------------------------------------------------------------------------")

# c. In ra tất cả các giá trị
print("c. Tất cả các giá trị:")
print(student.values())  # In ra đối tượng dict_values chứa tất cả các giá trị
# Cách khác: in ra dưới dạng danh sách
print(list(student.values()))  # Chuyển đổi dict_values thành list để hiển thị
print("------------------------------------------------------------------------------------")

# d. Cập nhật khóa học của Bruce từ Literature thành Foreign Languages
print("d. Cập nhật khóa học:")
print("Trước khi cập nhật:", student["Course"])
student["Course"] = "Foreign Languages"  # Gán giá trị mới cho khóa "Course"
print("Sau khi cập nhật:", student["Course"])
print("------------------------------------------------------------------------------------")

# e. Thêm thông tin còn thiếu: Address và Phone number (sử dụng hai cú pháp khác nhau)
print("e. Thêm thông tin còn thiếu:")
# Cú pháp 1: gán trực tiếp
student["Address"] = "123 College St, City"

# Cú pháp 2: sử dụng phương thức update()
student.update({"Phone number": "555-123-4567"})

print("Thông tin sinh viên sau khi cập nhật:")
for key, value in student.items():
    print(f"{key}: {value}")
print("------------------------------------------------------------------------------------")

# f. Xóa Sex và Hobby do chính sách bảo mật mới
print("f. Xóa trường Sex và Hobby:")
# Phương pháp xóa 1: sử dụng từ khóa del
del student["Sex"]  # Xóa khóa "Sex" và giá trị của nó

# Phương pháp xóa 2: sử dụng phương thức pop()
student.pop("Hobby")  # Xóa khóa "Hobby" và giá trị của nó

print("Thông tin sinh viên cuối cùng:")
for key, value in student.items():
    print(f"{key}: {value}")
```

```python
# bài tập 2
# 2. Tại một phòng khám thú y. Bạn là bác sĩ thú y tại một phòng khám thú y, và đây là một số thú cưng bạn hiện đang chăm sóc.
# pets = [{"name":"Toby", "animal type":"dog", "age":2},{"name":"Kitty", "animal type":"cat", "age":5},{"name":"Tiki", "animal type":"parrot", "age":1}]

# Khởi tạo danh sách các thú cưng
pets = [{"name":"Toby", "animal type":"dog", "age":2},
        {"name":"Kitty", "animal type":"cat", "age":5},
        {"name":"Tiki", "animal type":"parrot", "age":1}]

# a. Thêm một bệnh nhân mới (ngựa tên Sugar) vào danh sách
print("a. Thêm bệnh nhân mới:")
new_pet = {"name":"Sugar", "animal type":"horse", "age":4}
pets.append(new_pet)  # Thêm từ điển mới vào danh sách pets

# In ra danh sách sau khi thêm
print("Danh sách thú cưng sau khi thêm:")
for pet in pets:
    print(pet)
print("------------------------------------------------------------------------------------")

# b. In ra tất cả tên các loài động vật
print("b. Tên các loài động vật:")

# Cách 1: Sử dụng vòng lặp for qua các phần tử
print("Sử dụng vòng lặp for qua các phần tử:")
for pet in pets:  # Duyệt qua từng từ điển trong danh sách
    print("Tên:", pet['name'], ", Loài:", pet['animal type'])

print("\nSử dụng vòng lặp for qua các chỉ mục:")
# Cách 2: Sử dụng vòng lặp for qua các chỉ mục
for i in range(len(pets)):  # Duyệt qua các chỉ mục từ 0 đến len(pets)-1
    print("Tên:", pets[i]['name'], ", Loài:", pets[i]['animal type'])
print("------------------------------------------------------------------------------------")

# c. Thêm một mục cho biết tất cả các loài động vật hiện đang ở phòng khám
# Sử dụng kiểu dữ liệu set để lưu trữ các loài động vật duy nhất
print("c. Tất cả các loài động vật hiện có tại phòng khám:")

# Tạo một set để lưu trữ các loài động vật (không trùng lặp)
animal_types = set()  # Sử dụng kiểu dữ liệu set vì nó chỉ lưu các giá trị duy nhất

# Thêm loài của từng thú cưng vào set
for pet in pets:
    animal_types.add(pet["animal type"])

print("Các loài động vật hiện có tại phòng khám:", animal_types)
# Hoặc in ra dưới dạng danh sách
print("Danh sách các loài động vật:", list(animal_types))
```

```python
# bài tập 3
# 3. Nước ép! Bạn sở hữu một quầy nước ép và cần theo dõi lượng nước ép và doanh số bán hàng.


# a. Tạo danh sách các từ điển chứa thông tin về nước ép
juices = [
    {"flavor": "cam", "price": 25000, "color": "cam"},
    {"flavor": "chanh", "price": 20000, "color": "xanh lá"},
    {"flavor": "lựu", "price": 30000, "color": "đỏ"}
]

print("a. Danh sách các loại nước ép ban đầu:")
for juice in juices:
    print(juice)
print("------------------------------------------------------------------------------------")

# b. Thêm mục 'in shop' với giá trị Boolean cho mỗi loại nước ép
print("b. Thêm trạng thái 'in shop' cho mỗi loại nước ép:")
for juice in juices:
    juice["in shop"] = True  # Thêm khóa 'in shop' với giá trị True

# In ra danh sách sau khi thêm trạng thái
for juice in juices:
    print(juice)
print("------------------------------------------------------------------------------------")

# c. Thêm nước ép nho mới vào danh sách
print("c. Thêm nước ép nho vào danh sách:")
grape_juice = {"flavor": "nho", "price": 28000, "color": "tím", "in shop": True}
juices.append(grape_juice)

# In ra danh sách sau khi thêm nước ép nho
for juice in juices:
    print(juice)
print("------------------------------------------------------------------------------------")

# d. Tính giá trung bình của một loại nước ép
print("d. Tính giá trung bình của nước ép:")

# Tính tổng giá của tất cả các loại nước ép
total_price = 0
for juice in juices:
    total_price += juice["price"]

# Tính giá trung bình
average_price = total_price / len(juices)

print("Tổng giá của tất cả các loại nước ép:", total_price, "VND")
print("Số loại nước ép:", len(juices))
print("Giá trung bình của một loại nước ép:", average_price, "VND")
```
